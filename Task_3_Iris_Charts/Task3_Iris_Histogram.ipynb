```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Iris.csv')  # or 'iris.data' if that’s the file

print(df.head())

plt.figure(figsize=(8, 6))
plt.hist(df['SepalLengthCm'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Number of Samples')
plt.grid(True)
plt.show()
```

```python
species_count = df['Species'].value_counts()

plt.figure(figsize=(8, 6))
species_count.plot(kind='bar', color='orchid', edgecolor='black')
plt.title('Count of Each Iris Species')
plt.xlabel('Species')
plt.ylabel('Number of Samples')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```
